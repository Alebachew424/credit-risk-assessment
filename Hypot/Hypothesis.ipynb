```python
# ============================================
# FIRST: LOAD YOUR CLEANED DATA
# ============================================

import pandas as pd
import numpy as np
from scipy import stats

# Load the cleaned data
df = pd.read_csv('C:/Users/AlexB/Desktop/credit_risk_assessment/data/processed/credit_data_cleaned.csv')

print("=" * 60)
print("DATA LOADED SUCCESSFULLY!")
print("=" * 60)
print(f"Dataset shape: {df.shape[0]:,} rows, {df.shape[1]} columns")
print(f"Default rate: {df['SeriousDlqin2yrs'].mean():.2%}")
print("=" * 60)

# ============================================
# NOW RUN THE HYPOTHESIS TESTING CODE
# ============================================

print("\n" + "=" * 60)
print("STATISTICAL VALIDATION OF EDA FINDINGS")
print("=" * 60)

results = []

# 1. Age Test
young = df[df['age'] < 35]['SeriousDlqin2yrs']
old = df[df['age'] >= 55]['SeriousDlqin2yrs']
t_stat, p_val = stats.ttest_ind(young, old)
results.append(['Age', f'{young.mean():.2%} vs {old.mean():.2%}', p_val, '✅' if p_val < 0.05 else '❌'])

# 2. Utilization Test
low = df[df['RevolvingUtilizationOfUnsecuredLines'] <= 0.3]['SeriousDlqin2yrs']
high = df[df['RevolvingUtilizationOfUnsecuredLines'] >= 0.8]['SeriousDlqin2yrs']
t_stat, p_val = stats.ttest_ind(low, high)
results.append(['Utilization', f'{low.mean():.2%} vs {high.mean():.2%}', p_val, '✅' if p_val < 0.05 else '❌'])

# 3. Delinquency Test
clean = df[df['NumberOfTime30-59DaysPastDueNotWorse'] == 0]['SeriousDlqin2yrs']
past_due = df[df['NumberOfTime30-59DaysPastDueNotWorse'] > 0]['SeriousDlqin2yrs']
t_stat, p_val = stats.ttest_ind(clean, past_due)
results.append(['Delinquency', f'{clean.mean():.2%} vs {past_due.mean():.2%}', p_val, '✅' if p_val < 0.05 else '❌'])

# 4. Income Test
low_income = df[df['MonthlyIncome'] <= df['MonthlyIncome'].quantile(0.33)]['SeriousDlqin2yrs']
high_income = df[df['MonthlyIncome'] >= df['MonthlyIncome'].quantile(0.67)]['SeriousDlqin2yrs']
t_stat, p_val = stats.ttest_ind(low_income, high_income)
results.append(['Income', f'{low_income.mean():.2%} vs {high_income.mean():.2%}', p_val, '✅' if p_val < 0.05 else '❌'])

# Display results
print("\n📊 Hypothesis Test Results:")
print("-" * 60)
print(f"{'Feature':15} {'Comparison':25} {'P-Value':12} {'Valid?'}")
print("-" * 60)

for feature, comparison, p_val, valid in results:
    print(f"{feature:15} {comparison:25} {p_val:.6f}   {valid}")

print("\n" + "=" * 60)
print("CONCLUSION: Features marked ✅ are STATISTICALLY VALID predictors")
print("These WILL be used in your credit risk model")
print("=" * 60)
```
