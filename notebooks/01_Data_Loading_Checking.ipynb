```python
print("\n" + "=" * 60)
print("TARGET VARIABLE DISTRIBUTION")
print("=" * 60)

target_counts = df['SeriousDlqin2yrs'].value_counts()
target_percent = df['SeriousDlqin2yrs'].value_counts(normalize=True) * 100

print(f"\n0 = Good Credit (No Default):  {target_counts[0]:,} ({target_percent[0]:.2f}%)")
print(f"1 = Bad Credit (Default):    {target_counts[1]:,} ({target_percent[1]:.2f}%)")
print(f"\n Overall Default Rate: {df['SeriousDlqin2yrs'].mean():.2%}")
```

```python
print("\n" + "=" * 60)
print("MISSING VALUES ANALYSIS")
print("=" * 60)

missing_count = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing Percentage': missing_percent
})

missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False)

if len(missing_df) > 0:
    print("\n Columns with missing values:")
    print(missing_df)
else:
    print("\n No missing values found!")

print(f"\n Total missing values in dataset: {df.isnull().sum().sum():,}")
```

```python
print("\n" + "=" * 60)
print("DATA QUALITY CHECKS")
print("=" * 60)

# Check for duplicates
duplicates = df.duplicated().sum()
print(f"\n Duplicate rows: {duplicates}")

# Check for negative ages
negative_age = (df['age'] < 0).sum()
print(f" Negative ages: {negative_age}")

# Check for age = 0 (invalid)
zero_age = (df['age'] == 0).sum()
print(f" Age = 0 (invalid): {zero_age}")

# Check for age > 100
old_age = (df['age'] > 100).sum()
print(f" Age > 100 years: {old_age}")

# Check for utilization > 1
high_util = (df['RevolvingUtilizationOfUnsecuredLines'] > 1).sum()
print(f" Utilization > 100%: {high_util}")

# Check for utilization = 0
zero_util = (df['RevolvingUtilizationOfUnsecuredLines'] == 0).sum()
print(f" Utilization = 0%: {zero_util}")

# Check for negative debt ratio
neg_debt = (df['DebtRatio'] < 0).sum()
print(f" Negative debt ratio: {neg_debt}")

# Check for zero or negative income
bad_income = (df['MonthlyIncome'] <= 0).sum()
print(f" Zero or negative income: {bad_income}")

# Check for 98 values (possible placeholder)
past_due_98 = (df['NumberOfTime30-59DaysPastDueNotWorse'] == 98).sum()
print(f" Past due = 98 (possible placeholder): {past_due_98}")

# Check for 98 in other past due columns
past_due_60_98 = (df['NumberOfTime60-89DaysPastDueNotWorse'] == 98).sum()
past_due_90_98 = (df['NumberOfTimes90DaysLate'] == 98).sum()
print(f" Past due 60-89 days = 98: {past_due_60_98}")
print(f" Past due 90+ days = 98: {past_due_90_98}")
```
