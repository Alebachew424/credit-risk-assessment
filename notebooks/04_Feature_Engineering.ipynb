```python
"""
CREDIT RISK ASSESSMENT - FEATURE ENGINEERING
File: 04_Feature_Engineering.py
"""

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("=" * 60)
print("FEATURE ENGINEERING - CREDIT RISK ASSESSMENT")
print("=" * 60)

# Load cleaned data
print("\n Loading cleaned data...")
df = pd.read_csv('C:/Users/AlexB/Desktop/credit_risk_assessment/data/processed/credit_data_cleaned.csv')
print(f" Loaded: {df.shape[0]:,} rows, {df.shape[1]} columns")
```

```python
# ============================================
# CREATE NEW FEATURES
# ============================================

print("\n" + "=" * 60)
print("CREATING NEW FEATURES")
print("=" * 60)
```

```python
# 1. Total past due count
past_due_cols = ['NumberOfTime30-59DaysPastDueNotWorse', 
                 'NumberOfTime60-89DaysPastDueNotWorse',
                 'NumberOfTimes90DaysLate']
df['TotalPastDue'] = df[past_due_cols].sum(axis=1)
print("  Created: TotalPastDue")
```

```python
# 2. Any delinquency indicator
df['HasDelinquency'] = (df['TotalPastDue'] > 0).astype(int)
print("  Created: HasDelinquency")
```

```python
# 3. Severe delinquency indicator (90+ days)
df['SevereDelinquency'] = (df['NumberOfTimes90DaysLate'] > 0).astype(int)
print("  Created: SevereDelinquency")
```

```python
# 4. Recent delinquency (30-59 days)
df['RecentDelinquency'] = (df['NumberOfTime30-59DaysPastDueNotWorse'] > 0).astype(int)
print("  Created: RecentDelinquency")
```

```python
# 5. High utilization indicator
df['HighUtilization'] = (df['RevolvingUtilizationOfUnsecuredLines'] > 0.8).astype(int)
print("  Created: HighUtilization")
```

```python
# 6. Very high utilization (>0.9)
df['VeryHighUtilization'] = (df['RevolvingUtilizationOfUnsecuredLines'] > 0.9).astype(int)
print("  Created: VeryHighUtilization")
```

```python
# 7. Zero utilization indicator
df['ZeroUtilization'] = (df['RevolvingUtilizationOfUnsecuredLines'] == 0).astype(int)
print("  Created: ZeroUtilization")
```

```python
# 8. Has dependents indicator
df['HasDependents'] = (df['NumberOfDependents'] > 0).astype(int)
print("  Created: HasDependents")
```

```python
# 9. Many dependents (3+)
df['ManyDependents'] = (df['NumberOfDependents'] >= 3).astype(int)
print("  Created: ManyDependents")
```

```python

# 10. Income groups (already created in EDA)
# Keep the existing IncomeGroup

# 11. Age groups (already created in EDA)
# Keep the existing AgeGroup
```

```python
# 12. Age squared (for non-linear relationship)
df['AgeSquared'] = df['age'] ** 2
print("  Created: AgeSquared")

# 13. Log of income (to handle skewness)
df['LogIncome'] = np.log1p(df['MonthlyIncome'])
print("  Created: LogIncome")

# 14. Utilization per credit line
df['UtilPerCreditLine'] = df['RevolvingUtilizationOfUnsecuredLines'] / (df['NumberOfOpenCreditLinesAndLoans'] + 1)
print("  Created: UtilPerCreditLine")

# 15. Debt to income ratio (already exists as DebtRatio)
# Keep as is

# 16. Real estate to total loans ratio
df['RealEstateRatio'] = df['NumberRealEstateLoansOrLines'] / (df['NumberOfOpenCreditLinesAndLoans'] + 1)
print("  Created: RealEstateRatio")

print(f"\n Total features after engineering: {len(df.columns)}")
```

```python
# ============================================
# REVIEW NEW FEATURES
# ============================================

print("\n" + "=" * 60)
print("REVIEW NEW FEATURES")
print("=" * 60)

new_features = ['TotalPastDue', 'HasDelinquency', 'SevereDelinquency', 
                'RecentDelinquency', 'HighUtilization', 'VeryHighUtilization',
                'ZeroUtilization', 'HasDependents', 'ManyDependents',
                'AgeSquared', 'LogIncome', 'UtilPerCreditLine', 'RealEstateRatio']

print("\n New features summary:")
for feature in new_features:
    print(f"\n{feature}:")
    print(f"   Mean: {df[feature].mean():.4f}")
    print(f"   Std: {df[feature].std():.4f}")
    print(f"   Min: {df[feature].min():.4f}")
    print(f"   Max: {df[feature].max():.4f}")

    
```

```python
# ============================================
# CORRELATION WITH TARGET (UPDATED)
# ============================================

print("\n" + "=" * 60)
print("UPDATED CORRELATIONS WITH TARGET")
print("=" * 60)

# First, check if target column exists
print(f"Target column exists: {'SeriousDlqin2yrs' in df.columns}")

# Select features for modeling (exclude categorical groups)
feature_cols = [col for col in df.columns if col not in ['AgeGroup', 'UtilGroup', 'IncomeGroup', 'DebtGroup']]

# Calculate correlations (include target)
correlations_new = df[feature_cols].corr()['SeriousDlqin2yrs'].drop('SeriousDlqin2yrs').sort_values(ascending=False)

print("\n Top 15 features correlated with default:")
print("-" * 50)
for i, (feature, corr) in enumerate(correlations_new.head(15).items()):
    print(f"   {i+1:2}. {feature:35} {corr:+.4f}")
```
